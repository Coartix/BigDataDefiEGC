```python
import pandas as pd

actor_blockchain = pd.read_csv('timeseries/blockchain_by_actor.csv', index_col=0)


# Remove lines where year = 0
actor_blockchain = actor_blockchain[actor_blockchain['year'] != 0]


# combine "year", "month", and "day" columns into a single datetime column
actor_blockchain['date'] = pd.to_datetime(actor_blockchain[['year', 'month', 'day']])

# drop the original "year", "month", and "day" columns
actor_blockchain.drop(columns=['year', 'month', 'day'], inplace=True)

actor_blockchain
```

```python
# Change the nb_received and nb_transactions and nb_spent and sum_fee columns to integers
actor_blockchain = actor_blockchain.astype({'nb_received': 'int64', 'nb_transactions': 'int64', 'nb_spent': 'int64', 'sum_fee': 'int64'})
actor_blockchain
```

```python
actor_blockchain.set_index('identity', inplace=True)
actor_blockchain = actor_blockchain.groupby([actor_blockchain.index, 'date']).sum()
actor_blockchain
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
look = ['received']

plt.plot(actor_blockchain.loc['Paymium.com'].index, actor_blockchain.loc['Paymium.com'][look], label='Paymium.com')
plt.plot(actor_blockchain.loc['HaoBTC.com'].index, actor_blockchain.loc['HaoBTC.com'][look], label='HaoBTC.com')
plt.plot(actor_blockchain.loc['OKCoin.com'].index, actor_blockchain.loc['OKCoin.com'][look], label='OKCoin.com')
plt.plot(actor_blockchain.loc['BTCC.com'].index, actor_blockchain.loc['BTCC.com'][look], label='BTCC.com')
plt.plot(actor_blockchain.loc['Poloniex.com'].index, actor_blockchain.loc['Poloniex.com'][look], label='Poloniex.com')
plt.legend()
plt.show()


```
